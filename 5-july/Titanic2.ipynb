## Activation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,6))
```

```python
x= np.linspace(-5,5)
x
```

#### sigmoid

```python
def sigm(x):
    return 1/(1+np.exp(-x))
plt.plot(x, sigm(x))
plt.title("sigmoid activation function")
plt.show()
```

```python
def tnah(x):
    return np.tanh(x)
plt.plot(x, tnah(x))
plt.title("tnaH activation function")
plt.show()
```

```python
def relu(x):
    list = []
    for i in x:
        if i>=0:
            list.append(i)
        else:
            list.append(0)
    return list
plt.plot(x, relu(x))
plt.title("Rectified linear Unit activation function")
plt.show()
```

```python
def lrelu(x):
    list = []
    for i in x:
        if(i>=0):
            list.append(i)
        else:
            list.append(0.01*i)
    return list

plt.plot(x, lrelu(x))
plt.title("Leaky Rectified linear Unit activation function")
plt.show()
```

```python
def softm(x):
    return(np.exp(x))/np.sum(np.exp(x))

plt.plot(x, lrelu(x))
plt.title("Softmax activation function")
plt.show()
```
