```python
import pandas as pd
df = pd.read_csv('spam.csv')
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.text, df.label_num)
```

```python
x_train
```

```python
y_test
```

```python
df.text
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer()
```

```python
x_train_v = v.fit_transform(x_train.values)
```

```python
x_train_v
```

```python
x_train_v.toarray()[:2]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
model.fit(x_train_v, y_train)
```

```python
mail= ['Subject: hello']
mail_v = v.transform(mail)
model.predict(mail_v)
# 0- no_spam  , 1- spam
```

```python
mail= ['Congrats! Get ₹50,000 ke benefits with UPI waala Credit Card']
mail_v = v.transform(mail)
x = model.predict(mail_v)

if(x==0):print('not a spam')
else: print('spam mail')
```

```python
mail = input('enter subject')
mail_v = v.transform(mail)
x = model.predict(mail_v)

if(x==0):print('not a spam')
else: print('spam mail')
```
